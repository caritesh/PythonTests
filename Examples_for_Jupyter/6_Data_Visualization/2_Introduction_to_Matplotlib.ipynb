### Step 1: Import the Required Libraries and Load the Dataset

- Import pandas and matplotlib.pyplot libraries
- Load the HousePrices.csv file into a DataFrame and parse the 'date' column as datetime and set it as an index
- Sort the DataFrame by index
- Display the first 5 rows


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../../Datasets/HousePrices.csv',parse_dates=['date'],index_col=['date'])
df.sort_index(inplace=True)
```

```python
df.head(2)
```

__Observation__

- This is the head of the dataset.

### Step 2: Plot Various Graphs

Here, we will plot the various kinds of graphs used for visualization.


### Step 2.1: Plot the Line Plot

- Create a line plot of the __price__ column using the .plot method


```python
plt.plot(df['price'])
plt.show()
```

### Step 2.2: Plot the Box Plot

- Create a box plot of the 'price' column using .boxplot method

```python
plt.boxplot(df['price'])
plt.show()
```

### Step 2.3: Plot the Histogram
- Create a histogram of the 'sqft_living' column using the .hist method

```python
plt.hist(df['sqft_living'])
plt.show()
```

### Step 2.4: Plot the Scatter Plot
- Create a scatter plot with using .scatter method, where'price' on the x-axis using .xlabel and 'sqft_living' on the y-axis using .ylabel method. 
- We can also specify the title using .title method. 
 

```python
plt.xlabel('price')
plt.ylabel('sqft_living')
plt.title('Scatter Plot')
plt.scatter(x=df['price'],y = df['sqft_living'])
plt.show()
```

### Step 2.5: Plot the Bar Plot
- Create a bar plot using a dictionary

```python
d = {'a':10,'b':20,'c':13}
```

```python
plt.bar(x = d.keys(), height = d.values())
```

### Step 2.6: Plot the Pie Chart
- Create a pie chart using the same dictionary

```python
plt.pie(x = d.values(), labels = d.keys())
```

## Step 3: Customize a Line Plot

- Create a customized line plot of the 'price' column
- Set the figure size, dpi, x-axis label, y-axis label, title, and legend


```python
plt.plot(df['price'])
```

__Observation__

- This data is currently
too cumbersome to look at. 




Let's modify the properties of a figure
and save it.

- We will use dpi to display pixels per inch, which will
increase the resolution of the graph.

```python
plt.figure(figsize=(15,10),dpi = 100)
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Line Plot of Prices')
plt.legend()
plt.plot(df['price'])
```

__Observation__

- Now, the graph is enlarged.

### Step 4: Save the Graph in Local Directory

- The graph represented above can be saved in the local directory. 

- We will remove the legend( ) method and apply the .savfig( ) method to save this graph. 

```python
plt.figure(figsize=(15,10),dpi = 100)
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Line Plot of Prices')
plt.plot(df['price'])
plt.savefig('Prices.png')
```
